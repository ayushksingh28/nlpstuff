### Vectorization In Python

```python
x0, x1, x2 = 1., 2., 3.
bias, w1, w2 = 0.1, 0.3, 0.5

x = [x0, x1, x2]
w = [bias, w1, w2]
```

```python
def forloop(x, w):

  z = 0.
  for i in range(len(x)):
    z += x[i] * w[i]

  print(z)
```

### list comprehension

```python
def listcomp(x, w):
  z = sum(x_i *w_i for x_i, w_i in zip(x,w))
  print(z)
```

### use of numpy

```python
import numpy as np
x_vec, w_vec = np.array(x), np.array(w)
def vectorized(x_vec, w_vec):
  z = x_vec.dot(w_vec)
  print(z)

```

```python
%timeit -r 100 -n 10 forloop(x, w)
```

```python
%timeit -r 100 -n 10 listcomp(x, w)
```

```python
%timeit -r 100 -n 10 vectorized(x_vec, w_vec)
```
